```python
import pandas as pd
```

```python
df = pd.read_csv('/content/all_bovespa.csv', delimiter=',')
df
```

```python
# buscar todos os campos que comecem com preco
reg = df.filter(regex='^preco')
reg
```

```python
# retornar dados onde o campo finaliza com 'os'
reg = df.filter(regex='os$')
reg
```

```python
reg = df.filter(regex='preco[_].')
reg
```

```python
reg = df.filter(regex='.[_].')
reg
```

```python
# teste do replace
df.replace('3', '000', regex=True)
```
