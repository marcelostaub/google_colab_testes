```python
import pandas as pd
```

```python
#value count
df_prouni['curso_busca'].value_counts()
```

```python
#max
df_prouni['mensalidade'].max()
```

```python
#valor medio por estado
df_prouni.groupby(['uf_busca'])['mensalidade'].agg(['sum'])
```

```python
#valor medio por estado
df_prouni.groupby(['uf_busca'])['mensalidade'].agg(['mean'])
```

```python
#valor medio por curso por estado
df_prouni.groupby(['uf_busca', 'curso_busca'])['mensalidade'].agg(['mean'])
```

```python
#valor medio por curso e estado, colocando em ordem decrescente
df_prouni.groupby(['uf_busca', 'curso_busca'])['mensalidade'].agg(['mean']).sort_values(by=['uf_busca', 'mean'], ascending=False)
```

```python
#criar função - receberá registros (rows) de uma consulta e retornará no campo nova_bolsa o
#valor recebido multiplicado por 2, para todos os registros
def duplica_bolsa(rows):
  nova_bolsa = rows * 2
  return nova_bolsa
```

```python
df = df_prouni[['uf_busca', 'curso_busca','mensalidade','bolsa_integral_ampla' ]]
```

```python
df
```

```python
# chama a função duplica_bolsa, enviando como parâmetro o resultado da consulta anterior (dF)
df['nova_bolsa'] = df['bolsa_integral_ampla'].apply(duplica_bolsa)
```

```python
df
```

```python
# substituir as ocorrências de NaN co campo nova_bolsa por 1
df['nova_bolsa'] = df['nova_bolsa'].fillna(1)
```

```python
df
```

```python
#FUNÇÕES LAMBDA
# Fazendo a consulta a um campo e através da função lambda
# criar um novo campo recebendo o valor do campo consultado
# multiplicado por 2:
df['nova_bolsa_lambda'] = df['bolsa_integral_ampla'].apply(lambda x: x * 2 )
```

```python
df
```

```python
# ASSIGN
#criar um novo campo com o método assign e lambda
df = df.assign(nova_bolsa_lambda2 = lambda x: (x['bolsa_integral_ampla'] * 2))
```
