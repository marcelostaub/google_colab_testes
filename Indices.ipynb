```python
import pandas as pd
```

```python
df_acoes = pd.read_csv('/content/all_bovespa.csv', sep=',')
df_acoes
```

```python
#consulta acções Itaú:
df_itau = df_acoes[df_acoes['sigla_acao'] == 'ITUB4']
df_itau
```

"""# LOC
Como comando **loc** podemos pegar todos os registros de uma determianda coluna pelo nome da coluna.
O primeiro parametro indica quais as linhas que desejamos retornar e o segundo qual coluna.
"""

```python
#loc[ini_reg:final_reg,'coluna']
df1 = df_itau.loc[:,'data_pregao']  # retorna todos os registros da coluna 'data_pregao'
df1
```

```python
#fazer um slice, pgando apenas algumas linhas (registros)
df2= df_itau.loc[455:3776,'data_pregao']
df2
```

```python
# fazer a consulta retornando mais colunas
df3 = df_itau.loc[455:3776,['data_pregao', 'preco_abertura', 'preco_fechamento']]
df3
```

"""# Iloc
O comando iloc (integer location) faz a mesma coisa que o loc porem você só consegue utilizar os indices, tanto para as linhas quanto para as colunas.
"""

```python
# consulta os registros da coluna 'data_pregao' (índeice = 1)
df1 = df_itau.iloc[:, 1]
df1
```

```python
#acessando todos os dados do primeiro registro (primeira linha) do dataframe
df2 = df_itau.iloc[0]
df2
```

```python
# consultando a 1ª linha e todas as colunas do dataframe
# vai retornar os mesmos dados da consulta anterior, mas organizados de outra forma
df2 = df_itau.iloc[0:1,:]
df2
```

```python
#todas as linhas e 2 colunas (2:4 = colunas 2 e 3)
df3 = df_itau.iloc[:, 2:4]
df3
```

```python
# 2 linhas (1:3 = linhas 1 e 2) e 2 colunas (2:4 = colunas 2 e 3)
df4 = df_itau.iloc[1:3, 2:4]
df4
```

"""Mas o nosso dataframe do itau está com index estranhos?
O que acontece é que após a consulta para retornar as ações do Itaú, o dataframe permaneceu com os indexes originais.

Vamos ajustar para colocar os indices começando novamente do zero
"""

```python
# comando reset_index()
# Com o comando reset_index() é possível reinicar o índice do dataframe_
# _sendo incrementado de 1 em 1, a partir do 0 até o número máximo de registros
df_itau.reset_index()
```

```python
# Após executar o reset_index() uma nova coluna, chamada 'index', é criada automaticamente
# Esta nova coluna 'index' armazena o valor dos índices originais do dataframe
# Para que esta nova coluna não seja criada, basta executar o comando de reset passando
# o parâmetro drop=True
df_itau.reset_index(drop=True)
```

```python
#pegar as 10 perimeiras linhas da primeira coluna
#isso não vai funcionar
df_itau.loc[1:10,'data_pregao']
```

```python
# A consulta anterior não retornou os dados desejados porque porque a alteração dos índices
# não é feita do data frame original, pois o pandas não altera o original por default
df_itau
```

```python
# Temos que inserir um argumeto para que ocorra essa alteração de índices no dataframe original
# Parâmetro inplace, com valor True, deve ser passado para a função
df_itau.reset_index(inplace=True, drop=True)
```

```python
df_itau
```

```python
# Agora sim a consulta deve funcionar
df_itau.loc[1:10,'data_pregao']
```

```python
# Se quisermos colocar outra coluna como índice do dataframe podemos fazer isso da seguinte maneira
# Criando um objeto de dados para teste
empresa = {'email': ['jose@email.com', 'maria@email.com', 'antonio@email.com'],
   'departamento': ['marketing','financeiro','contabilidade'],
   'nome': ['Jose','Ana','Antonio']}
```

```python
# Cria um dataframe de teste a partir do objeto empresa
df_empresa = pd.DataFrame(empresa)
df_empresa
```

```python
# Setando o campo email como índice do dataframe
# comando set_index
df_empresa.set_index('email')
```

```python
df_empresa
```

```python
# De novo, ao visualizar o dataframe, verificamos que não foi alterado pois essa mudança
# de índice não é feita no dataframe original, pois o pandas não altera o original por default.
# Temos que inserir o argumeto inplace=True para que ocorra a alteração do índicde no dataframe.
df_empresa.set_index('email', inplace=True)

df_empresa
```

```python
# Fazendo uma consulta pelo índice novo:
df_empresa.loc['maria@email.com']
```

```python
# Revertendo a alteração no índice
df_empresa.reset_index()
```

```python
df_empresa
```

```python
# Ao consultar df_empresa, vemos que continuou com o e-mail como índice
# porque não foi passado o parâmetro inplace=True
df_empresa.reset_index(inplace=True)

df_empresa
```
