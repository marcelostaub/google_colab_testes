```python
import pandas as pd
# DADOS DUMMIES
#ler dados
df = pd.read_csv("http://bit.ly/kaggletrain")
df
```

```python
# get_dummies - transforma o dado categórico em lógico
pd.get_dummies(df.Sex)
```

```python
#criar embarque dummy
emb = pd.get_dummies(df.Embarked)
emb
```

```python
#concatenar no df
df = pd.concat([df,emb],axis=1 )
df
```

```python
# Fazer a concatenação do dado dummie diretamente no dataframe original e eliminar o primeiro campo dummie criado
# para que só fique o male ou femal na tabela e não os dois campos, pois um deles já basta
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)
df
```

```python
# CHECANDO OUTLIERS

#Outliers
df = pd.read_csv("//content/weight-height.csv", delimiter=',')
df
```

```python
# importa biblioteca mathplotlib para criar gráficos
from matplotlib import pyplot as plt
```

```python
# criar um histograma dos dados referentes ao peso
plt.hist(df.Weight, bins=30, rwidth=0.8)
```

```python
# obter o zscore para cada registro (zscore = valor_registro - valor_médio)/desvio_padrão
df['zscore'] = ( df.Weight - df.Weight.mean() ) / df.Weight.std()
df
```

```python
# filtra os zscore maiores que +2 e também os zscore menores que -2
df = df[(df.zscore< -3) | (df.zscore>3)]
df
```

```python
# filtra os zscore maiores que +3 e menores que -3
df = df[(df.zscore< -3) | (df.zscore>3)]
df
```
