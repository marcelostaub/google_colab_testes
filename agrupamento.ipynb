```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/cursos-prouni.csv")
df
```

"""pandas.crosstab(index, columns, values=None, rownames=None, colnames=None, \
                aggfunc=None, margins=False, margins_name: str = 'All', \
                dropna: bool = True, normalize=False)

1.  index: Valores para agrupar nas linhas
2.  columns: Valores para agrupar nas colunas
3.  values: Dados a serem agregados
4.  aggfunc: Função utilizada para agregar
5.  margins: Retorna o subtotal das linhas e colunas
6.  normalize: Divide os valores pela total
"""

```python
#Criando pd.crosstab com tipo de curso por estado.
pd.crosstab(df.uf_busca,df.nome)
```

```python
#Criando pd.crosstab com a cursos por UF com subtotal por linha e coluna
pd.crosstab(df.uf_busca,df.nome, margins=True)
```

```python
#Adicionar níveis
pd.crosstab(df.uf_busca, [df.nome,df['turno']], margins=True)
```

```python
#agrupando por cursos nas linhas e estados nas colunas
pd.crosstab([df.nome,df['turno']], [df.uf_busca] , margins=True)
```

```python
#Quantidade de bolsa integral ampla por cursos
#usar o parametro values
pd.crosstab([df.nome], [df.uf_busca] , values=df.bolsa_integral_ampla, aggfunc= np.sum, margins=True)
```

```python
#agrupar por produto
df_tipo_curso = df.groupby('nome')
df_tipo_curso
```

```python
#quantidade de grupos
df_tipo_curso.ngroups
```

```python
df.groupby('uf_busca').ngroups
```

```python
#atributo groups para mostrar os grupos (monta uma lista de registros por grupo)
df.groupby('uf_busca').groups
```

```python
#atributo size para mostrar o tamanho do grupos
df.groupby('uf_busca').size()
```

```python
#atributo get_group para filtrar um grupo especifico o tamanho do grupos
df.groupby('uf_busca').get_group('RS')
```

```python
#média
df.groupby('uf_busca').mean(numeric_only=True)
```

```python
# agrupar por mais de um campo e somando
df.groupby(['uf_busca', 'nome', ])['bolsa_integral_ampla'].sum()
```
